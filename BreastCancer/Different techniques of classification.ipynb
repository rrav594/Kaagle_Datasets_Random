```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%config BackendInline.figure_format = 'retina'
```

```python
data = pd.read_csv('data.csv')
```

```python
data.describe().T
```

```python
data.drop(['Unnamed: 32','id'],axis =1,inplace =True)
```

```python
data.info()
```

```python
data.diagnosis = data['diagnosis'].map({'M':1,'B':0})
```

```python
data.head()
```

```python
data.columns
```

```python
data_y = data.pop('diagnosis')
```

```python
data_x = data
```

```python
data_x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_x, data_y, test_size = 0.3, random_state = 14)
```

```python
from sklearn.feature_selection import SelectFromModel
```

```python
### Decision Tree Classifier
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf_dt = DecisionTreeClassifier()
clf_dt.fit(X_train,y_train)
```

```python
y_pred = clf_dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
### AdaBoostClassifier
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
clf_ad = AdaBoostClassifier()
params = {'n_estimators':[25,50,75,100,130,150,180,500,1000],'learning_rate':[0.2,0.5,0.8,1,1.4,1.7,2.5]}
rand = RandomizedSearchCV(clf_ad,param_distributions=params, scoring='precision')
```

```python
rand.fit(X_train,y_train)
```

```python
best_clf = rand.best_estimator_
```

```python
y_pred = best_clf.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import precision_score, recall_score, confusion_matrix, classification_report
```

```python
precision_score(y_test, y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.svm import SVC
```

```python
clf_svm = SVC()
params = {'C':[0.2,0.5,0.8,1,1.2,1.5,1.8,2.5,2.9,5],
        'kernel':['rbf','poly'],
          'gamma':[0.2,0.5,0.8,1.2,1.5,1.8,2.1,2.5,2.9],
          'degree':[1,2,3,4,5]
         }
rand = RandomizedSearchCV(estimator=clf_svm,param_distributions=params,scoring='accuracy',cv = 5)
```

```python
rand.fit(X_train,y_train)
```

```python
best_svm = rand.best_estimator_
best_svm
```
